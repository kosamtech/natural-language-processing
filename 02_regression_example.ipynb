```python
import zipfile
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
with zipfile.ZipFile("/content/airfoil+self+noise.zip") as thezip:
    thezip.extractall()
```

```python
df = pd.read_csv("/content/airfoil_self_noise.dat", sep="\t", header=None)
df.head()
```

```python
df.shape
```

```python
data = df[[1,2,3,4]].values
data
```

```python
targets = df[5].values
```

```python
targets
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data, targets, test_size=0.33)
```

```python
Xtrain.shape, Xtest.shape
```

```python
model = LinearRegression()
model.fit(Xtrain, Ytrain)
```

```python
model.score(Xtrain, Ytrain)
```

```python
model.score(Xtest, Ytest)
```

```python
predictions = model.predict(Xtest)
```

```python
predictions
```

```python
model2 = RandomForestRegressor()
model2.fit(Xtrain, Ytrain)
```

```python
model2.score(Xtrain, Ytrain)
```

```python
model2.score(Xtest, Ytest)
```
