```python
import numpy as np
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
```

### Using Scikit Learn Random Forest Classifier

```python
data = load_breast_cancer()
```

```python
type(data)
```

```python
data.keys()
```

```python
data.data.shape
```

```python
data.target.shape
```

```python
data.target_names
```

```python
data.feature_names
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(data.data, data.target, test_size=0.33)
```

```python
model = RandomForestClassifier()
model.fit(Xtrain, Ytrain)
```

```python
model.score(Xtrain, Ytrain)
```

```python
model.score(Xtest, Ytest)
```

```python
prediction = model.predict(Xtest)
```

```python
prediction
```

```python
prediction.shape
```

```python
prediction == Ytest
```

```python
N = len(Ytest)
np.sum(prediction == Ytest) / N
```

### Using Deep Learning Solution

```python
from sklearn.neural_network import MLPClassifier
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
Xtrain2 = scaler.fit_transform(Xtrain)
Xtest2 = scaler.fit_transform(Xtest)
```

```python
Xtrain2.shape, Xtest2.shape
```

```python
model = MLPClassifier(max_iter=500)
model.fit(Xtrain2, Ytrain)
```

```python
model.score(Xtrain2, Ytrain)
```

```python
model.score(Xtest2, Ytest)
```
